```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.espn.com/nba/team/schedule/_/name/sac'
```

```python
page_response = requests.get(url, timeout=5)
page_response
```

```python
page_content = BeautifulSoup(page_response.content, "html.parser")
```

```python
page_content.prettify()
```

```python
page_content.find('td', attrs={'class': 'Table2__td'}).find('span').text
```

```python
gamedays = []
```

```python
for ptag in page_content.find_all('td', attrs={'class': 'Table2__td'}):
    # prints the p tag content
    try:
        gamedays.append(ptag.find('span').text)
    except AttributeError:
        pass
```

```python
gamedays
```

```python
gamedays = gamedays[5:]
gamedays
```

```python
gamedays = gamedays[::3]
gamedays
```

```python
page_content.find('div', attrs = {'class': 'flex items-center opponent-logo'}).find_all('a')[1]
```

```python
opponent = []
```

```python
for ptag in page_content.find_all('div', attrs = {'class': 'flex items-center opponent-logo'}):
    try:
        opponent.append(ptag.find_all('a')[1].text)
    except AttributeError:
        pass
```

```python
opponent
```

```python
kings_schedule = pd.DataFrame(columns = ['Date', 'Opponent'])
```

```python
home_away = []
```

```python
page_content.find('div', attrs = {'class': 'flex items-center opponent-logo'}).find('span').text
```

```python
for ptag in page_content.find_all('div', attrs = {'class': 'flex items-center opponent-logo'}):
    try:
        home_away.append(ptag.find('span').text)
    except AttributeError:
        pass
```

```python
kings_schedule['Date'] = pd.Series(gamedays)
kings_schedule['Opponent'] = pd.Series(opponent)
kings_schedule['Home/Away'] = pd.Series(home_away)
kings_schedule
```

```python
kings_schedule = kings_schedule[['Date', 'Home/Away', 'Opponent']]
kings_schedule
```

```python
kings_schedule['Home/Away'] = kings_schedule['Home/Away'].str.replace('@', 'Away')
kings_schedule['Home/Away'] = kings_schedule['Home/Away'].str.replace('vs', 'Home')
```

```python
kings_schedule
```

```python
kings_schedule.to_csv('Kings Schedule.csv', index = False)
```
