```python
page_content = BeautifulSoup(page_response.content, "html.parser")
```

```python
def get_gamedays(page_content):
    gamedays = []
    for ptag in page_content.find_all('td', attrs={'class': 'Table2__td'}):
    # prints the p tag content
        try:
            gamedays.append(ptag.find('span').text)
        except AttributeError:
            pass
    gamedays = gamedays[5::3]
    return gamedays
```

```python
def get_home_away(page_content):    
    home_away = []
    for ptag in page_content.find_all('div', attrs = {'class': 'flex items-center opponent-logo'}):
        try:
            home_away.append(ptag.find('span').text)
        except AttributeError:
            pass
    return home_away
```

```python
def get_opponents(page_content):
    opponent = []
    for ptag in page_content.find_all('div', attrs = {'class': 'flex items-center opponent-logo'}):
        try:
            opponent.append(ptag.find_all('a')[1].text)
        except AttributeError:
            pass
    return opponent
```

```python
def input_process():
    team = '0'
    while True:
        try: 
            team = int(input('''0 - Atlanta Hawks
1 - Boston Celtics
2 - Brooklyn Nets
3 - Charlotte Hornets
4 - Chicago Bulls
5 - Clevand Cavaliers
6 - Dallas Mavericks
7 - Denver Nuggets
8 - Detroit Pistons
9 - Golden State Warriors
10 - Houston Rockets
11 - Indiana Pacers
12 - LA Clippers
13 - Los Angeles Lakers
14 - Memphis Grizzlies
15 - Miami Heat
16 - Milwaukee Bucks
17 - Minnesota Timberwolves
18 - New Orleans Pelicans
19 - New York Knicks
20 - Oklahoma City Thunder
21 - Orlando Magic
22 - Philedelphia 76ers
23 - Phoenix Suns
24 - Portland Trailblazers
25 - Sacramento Kings
26 - San Antonio Spurs
27 - Toronto Raptors
28 - Utah Jazz
29 - Washington Wizards
        '''))
            team = team_get[team]
            break
        except (ValueError, KeyError) as e:
            print('Please give a number between 0 and 29')
            continue
    return team
```

```python
def main(team):
    url = 'https://www.espn.com/nba/team/schedule/_/name/' + team
    page_response = requests.get(url, timeout=5)
    page_content = BeautifulSoup(page_response.content, "html.parser")
    gamedays = get_gamedays(page_content)
    home_away = get_home_away(page_content)
    opponent = get_opponents(page_content)
    team_schedule = pd.DataFrame(data={'Date':gamedays, 'Home/Away':home_away,'Opponent':opponent})
    team_schedule['Home/Away'] = team_schedule['Home/Away'].str.replace('@', 'Away')
    team_schedule['Home/Away'] = team_schedule['Home/Away'].str.replace('vs', 'Home')
    return team_schedule
```

```python
team = input_process()
cha_hornets = main(team)
cha_hornets
```

```python
toronto = input_process()
tor_raptors = main(toronto)
tor_raptors
```
